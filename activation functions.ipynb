```python
# Activation Functions in Neural Networks
# Sigmoid, Tanh, ReLU, Leaky ReLU, Softmax
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))  # classic formula
```

```python
# tanh function
def tanh(x):
    return np.tanh(x)  # numpy has it already
```

```python
# relu function (very common one)
def relu(x):
    return np.maximum(0, x)  # 0 if negative
```

```python
# leaky relu - small slope for negative x
def leaky_relu(x):
    alpha = 0.01  # just a small constant
    return np.where(x > 0, x, alpha * x)
```

```python
# softmax - used in last layer of classifiers
def softmax(x):
    # shift x to avoid big exponents
    shifted = x - np.max(x)
    exp_vals = np.exp(shifted)
    return exp_vals / np.sum(exp_vals)
```

```python
# preparing values for x-axis
x_vals = np.linspace(-10, 10, 100)
```

```python
# starting plots
plt.figure(figsize=(12, 8))
```

```python
# 1. Sigmoid
plt.subplot(2, 3, 1)
plt.plot(x_vals, sigmoid(x_vals), 'm')  # m = magenta
plt.title("Sigmoid")
plt.grid(True)
```

```python
# 2. Tanh
plt.subplot(2, 3, 2)
plt.plot(x_vals, tanh(x_vals), 'b')  # blue
plt.title("Tanh")
plt.grid(True)
```

```python
# 3. ReLU
plt.subplot(2, 3, 3)
plt.plot(x_vals, relu(x_vals), 'g')  # green
plt.title("ReLU")
plt.grid(True)
```

```python
# 4. Leaky ReLU
plt.subplot(2, 3, 4)
plt.plot(x_vals, leaky_relu(x_vals), 'r')  # red
plt.title("Leaky ReLU")
plt.grid(True)
```

```python
# 5. Softmax (just showing it on a small list)
some_vals = np.array([1.5, 2.0, 2.5, 3.0])
soft_vals = softmax(some_vals)
plt.subplot(2, 3, 5)
plt.bar(range(len(some_vals)), soft_vals, color='orange')
plt.title("Softmax Output")
plt.grid(True)
```

```python
# --- Input Values ---
x_vals = np.linspace(-10, 10, 500)

# Evaluate continuous functions
sig_vals = sigmoid(x_vals)
tanh_vals = tanh(x_vals)
relu_vals = relu(x_vals)
leaky_vals = leaky_relu(x_vals)

# Softmax demo on a fixed vector
soft_vec = np.array([1, 2, 3, 4, 5])
softmax_vals = softmax(soft_vec)

# --- Plotting ---
plt.figure(figsize=(10, 6))

# Plot continuous curves
plt.plot(x_vals, sig_vals, label='Sigmoid', color='magenta')
plt.plot(x_vals, tanh_vals, label='Tanh', color='blue')
plt.plot(x_vals, relu_vals, label='ReLU', color='green')
plt.plot(x_vals, leaky_vals, label='Leaky ReLU', color='red')

# Plot Softmax values as points
for i, val in enumerate(softmax_vals):
    plt.scatter(i, val, color='orange', label='Softmax' if i == 0 else "")
    plt.text(i, val + 0.02, f"{val:.2f}", ha='center', fontsize=8)

# Labels and formatting
plt.title("Activation Functions (All in One Plot)")
plt.xlabel("Input (x)")
plt.ylabel("Activation Output")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
